**DataFrame using Series**

```python
import pandas as pd
arr={"a":pd.Series([1,2,3,4,5]),"b":pd.Series([1,2,3,4,5])}
print(pd.DataFrame(arr))
```

**Arithemetic op**

```python
import pandas as pd
var=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
print(var,"\n")
var["Sum"]=var["A"]+var["B"]
print(var,"\n")

var["diff"]=var["A"]-var["B"]
print(var,"\n")
var["Acheck"]=var["A"]<=3
print(var,"\n")
var["Bcheck"]=var["B"]>=1
print(var,"\n")
```
